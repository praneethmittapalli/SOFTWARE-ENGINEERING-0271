```python
import csv
import matplotlib.pyplot as plt
import numpy as np

def quadratic_solver(a_coeff, b_coeff, c_coeff):
    delta = b_coeff**2 - 4 * a_coeff * c_coeff
    if delta >= 0:
        x1 = (-b_coeff + (delta**0.5)) / (2 * a_coeff)
        x2 = (-b_coeff - (delta**0.5)) / (2 * a_coeff)
        return x1, x2
    else:
        return None

def plot_quadratic_equation(a_coeff, b_coeff, c_coeff, solutions):
    x_values = np.linspace(-10, 10, 100)
    y_values = a_coeff * x_values**2 + b_coeff * x_values + c_coeff

    plt.plot(x_values, y_values, label=f'{a_coeff}x^2 + {b_coeff}x + {c_coeff}')

    if solutions:
        plt.scatter(solutions, [0, 0], color='red', marker='x', label='Solutions')

    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title('Quadratic Equation and Solutions')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

# Updated coefficients
new_a = 2
new_b = -5
new_c = 3

# Calculate solutions for the updated coefficients
new_solutions = quadratic_solver(new_a, new_b, new_c)

# Print the solutions
print("Solutions:", new_solutions)

# Plot the updated quadratic equation and solutions
plot_quadratic_equation(new_a, new_b, new_c, new_solutions)

# Write coefficients and solutions to a CSV file
csv_filename = "quadratic_equations.csv"
with open(csv_filename, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Coefficient a', 'Coefficient b', 'Coefficient c', 'Solution 1', 'Solution 2'])
    csv_writer.writerow([new_a, new_b, new_c, new_solutions[0] if new_solutions else None, new_solutions[1] if new_solutions else None])

print(f"Data written to {csv_filename}")
```
