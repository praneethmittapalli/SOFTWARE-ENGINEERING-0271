```python
import matplotlib.pyplot as plt
import numpy as np

def plot_weather(days_array, temperature, rainfall):
    plt.plot(days_array, temperature, label='Temperature')
    plt.plot(days_array, rainfall, label='Rainfall')
    plt.xlabel('Days')
    plt.ylabel('Value')
    plt.title('Weather Modeling with Quadratic Solution')
    plt.legend()
    plt.show()

# Updated values
new_parameters = [25, 11.8, 2.3, 25.0, 17.0]

# Unpack the updated values
new_days, new_temp_coeff, new_rain_coeff, new_init_temp, new_init_rain = new_parameters

# Calculate updated temperature and rainfall values
new_days_array = np.arange(1, int(new_days) + 1)
new_temperature = new_init_temp - new_temp_coeff * (new_days_array - 1) ** 2
new_rainfall = new_init_rain + new_rain_coeff * (new_days_array - 1) ** 2

# Plot the updated values
plot_weather(new_days_array, new_temperature, new_rainfall)
```
