```python
import matplotlib.pyplot as plt
import numpy as np

def generate_quadratic_equation(a_coeff, b_coeff, c_const, time_values):
    temperature_values = a_coeff * time_values**2 + b_coeff * time_values + c_const
    return temperature_values

def plot_quadratic_equation(time, temperature, additional_time, additional_temperature):
    plt.plot(time, temperature, label='Temperature Model', color='blue')
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.title('Weather Modeling with Quadratic Solution')

    plt.scatter(additional_time, additional_temperature, color='red', label='Additional Point')
    plt.legend()
    plt.show()

# Updated coefficients for the quadratic equation
new_a = 15.0  # coefficient for the quadratic term
new_b = 9.0   # coefficient for the linear term
new_c = 35.0  # constant term

# Generate time values
new_time_values = np.linspace(0, 10, 100)

# Calculate temperature values using the new coefficients
new_temperature_values = generate_quadratic_equation(new_a, new_b, new_c, new_time_values)

# Updated additional time value
new_additional_time = 4

# Calculate temperature for the additional time
new_additional_temperature = generate_quadratic_equation(new_a, new_b, new_c, np.array([new_additional_time]))

# Plotting the quadratic function with the updated values
plot_quadratic_equation(new_time_values, new_temperature_values, new_additional_time, new_additional_temperature)
```
