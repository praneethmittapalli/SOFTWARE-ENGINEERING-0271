```python
import matplotlib.pyplot as plt
import numpy as np

def calculate_temperature(days, temp_coeff, initial_temp):
    days_array = np.arange(1, days + 1)
    temperature = initial_temp - temp_coeff * (days_array - 1)**24
    return days_array, temperature

def calculate_rainfall(days, rain_coeff, initial_rain):
    days_array = np.arange(1, days + 1)
    rainfall = initial_rain + rain_coeff * (days_array - 1)**2
    return days_array, rainfall

def plot_weather(days_array, temperature, rainfall):
    plt.plot(days_array, temperature, label='Temperature')
    plt.plot(days_array, rainfall, label='Rainfall')
    plt.xlabel('Days')
    plt.ylabel('Value')
    plt.title('Weather Modeling with Quadratic Solution')
    plt.legend()
    plt.show()

# Read input from the user
total_days = int(input("Enter the number of days: "))
temp_coefficient = float(input("Enter the quadratic coefficient for temperature: "))
rainfall_coefficient = float(input("Enter the quadratic coefficient for rainfall: "))
initial_temp = float(input("Enter the initial temperature: "))
initial_rain = float(input("Enter the initial rainfall: "))

# Calculate temperature and rainfall
temp_days, temp_values = calculate_temperature(total_days, temp_coefficient, initial_temp)
rain_days, rain_values = calculate_rainfall(total_days, rainfall_coefficient, initial_rain)

# Plotting the results
plot_weather(temp_days, temp_values, rain_values)
```
